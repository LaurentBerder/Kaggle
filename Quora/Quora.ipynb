```python
path = "C:\Users\galdo\PycharmProjects\Kaggle\Quora\\"
train = pd.read_csv(path+"train.csv.zip", index_col=0, compression='zip')
train
```

```python
# Transform questions in vectors of words
train['tok1'] = train.apply(lambda row: nltk.word_tokenize(str(row['question1']).decode('utf-8').lower()), axis=1)
train['tok2'] = train.apply(lambda row: nltk.word_tokenize(str(row['question2']).decode('utf-8').lower()), axis=1)
```

```python
# Import stopwords to focus on important content of sentence.
stop_words = set(nltk.corpus.stopwords.words('english'))
```

```python
# Filter stopwords out of vectors
train['filtered1'] = train['tok1'].apply(lambda x: [word for word in x if word not in stop_words])
train['filtered2'] = train['tok2'].apply(lambda x: [word for word in x if word not in stop_words])
```

```python
# Import 
ps =  nltk.stem.porter.PorterStemmer()
```
